> Desafio 004 == Dissecando uma variável

```python
n = input('Digite alguma coisa:')
print('O que você digitou é numerico? ', n.isnumeric())
print('O que você digitou é uma letra? ', n.isalpha())
print('O que você digitou é uma letra e um número? ', n.isalnum())
print('O que você digitou está escrito em letras maiúsculas? ', n.isupper())
print('O que você digitou está escrito em letras minúsculas? ', n.islower())
print('O que você digitou está capitalizado? ', n.istitle())
print('O que você digitou só tem espaço ou está em branco? ', n.isspace())
```

> Desafio 007 == Média Aritimetica

```python
nota1 = float(input('Digite a nota do primeiro bimestre: '))
nota2 = float(input('Digite a nota do segundo bimestre: '))
nota3 = float(input('Digite a nota do terceiro bimestre: '))
nota4 = float(input('Digite a nota do quarto bimestre: '))
média = (nota1 + nota2 + nota3 + nota4)/4
if média > 5.99:
  print('O aluno passou de ano! e a sua média é igual a: {:.2f}'.format(média))
else:
  print('Infelismente o aluno vai ficar de recuperação!Pois a sua média foi: {}'.format(média))
```

> Desafio 018 == Seno, Cosseno e Tangente

```python
import math
angulo = float(input('Digite o ângulo que você deseja: '))
seno = math.sin(math.radians(angulo))
print('O ângulo de {} tem o seno de {:.2f}'.format(angulo, seno))
cosseno = math.cos(math.radians(angulo))
print('O ângulo de {} tem o cosseno de {:.2f}'.format(angulo, cosseno))
tangente = math.tan(math.radians(angulo))
print('O ângulo de {} tem o tangente de {:.2f}'.format(angulo, tangente))
```

> Desafio 019 == Sorteando um item na lista

```python
import random
a1 = input('Digite o nome do 1° aluno: ')
a2 = input('Digite o nome do 2° aluno: ')
a3 = input('Digite o nome do 3° aluno: ')
a4 = input('Digite o nome do 4° aluno: ')
lista = [a1, a2, a3, a4]
escolhido = random.choice(lista)
print('O aluno escolhido foi o {}'.format(escolhido))
```

Desafio 020 == Sorteando uma lista

```python
import random
p1 = input('Digite o nome da 1° pessoa: ')
p2 = input('Digite o nome da 2° pessoa: ')
p3 = input('Digite o nome da 3° pessoa: ')
p4 = input('Digite o nome da 4° pessoa: ')
lista = [p1, p2, p3, p4]
random.shuffle(lista)
print('A ordem de apresentação será: ')
print(lista)
```
