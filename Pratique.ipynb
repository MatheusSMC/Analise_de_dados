```python
#Qual é o seu nome?
nome = input("Digite seu nome: ")
print(f"Olá," , nome)
```

```python
valor = int(input("Digite o primeiro valor: "))
valor2 = int(input("Digite o segundo valor: "))
operação = input("Digite uma operação: (+ , -, *, /) ")
if operação == "+":
  print(nome , "o resultado da operação e:" , valor + valor2)
elif operação == "-":
  print(nome , "o resultado da operação e:" , valor - valor2)
elif operação == "*":
  print(nome , "o resultado da operação e:" , valor * valor2)
elif operação == "/":
  print(nome , "o resultado da operação e:" , valor / valor2)
else:
  print(nome , "Essa operação é inválida")
```
